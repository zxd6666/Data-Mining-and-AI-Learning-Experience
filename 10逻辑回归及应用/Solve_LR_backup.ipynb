## 1. 使用梯度下降法求解逻辑回归参数


\begin{equation*}
\theta_j = \theta_j - \frac{\partial J(\theta)}{\partial \theta_j}\\
=\theta_j + \alpha\frac{1}{m}\sum_{i=1}^m\left[ y^{(i)}-h_\theta\left(x^{(i)}\right)\right]\,x_j^{(i)} \\
h_\theta(x^{(i)}) = \frac{1}{1+e^{\sum_{j=1}^{n}-\theta_j x_j^{(i)}}}
\end{equation*}
$i$是样本点的标号，$j$是自变量个数的标号，在我们的案例中，$i$取1至731，$j$只取1和2(有常量参数）


```python
def sigmoid(x, theta_1, theta_2):                                                        
    z = (theta_1*x + theta_2).astype("float_")                                              
    return 1.0 / (1.0 + np.exp(-z)) 


def Cost(x, y, theta_1, theta_2):                                                                
    sigmoid_probs = sigmoid(x, theta_1, theta_2)                                        
    return -np.mean(y * np.log(sigmoid_probs)
                  + (1 - y) * np.log(1 - sigmoid_probs)) 

def gradient(x, y, theta_1, theta_2):                                                         
    sigmoid_probs = sigmoid(x, theta_1, theta_2)                                        
    return 1/len(y)*np.sum((y - sigmoid_probs)*x),1/len(y)*np.sum((y - sigmoid_probs)*1)                   

def hessian(x, y, theta_1, theta_2):                                                          
    sigmoid_probs = sigmoid(x, theta_1, theta_2)                                        
    d1 = np.sum((sigmoid_probs * (1 - sigmoid_probs)) * x * x)                  
    d2 = np.sum((sigmoid_probs * (1 - sigmoid_probs)) * x * 1)                  
    d3 = np.sum((sigmoid_probs * (1 - sigmoid_probs)) * 1 * 1)                  
    H = np.array([[d1, d2],[d2, d3]])                                           
    return H


```

```python
def gradient_nt(x, y, theta_1, theta_2):                                                         
    sigmoid_probs = sigmoid(x, theta_1, theta_2)                                        
    return np.array([[np.sum((y - sigmoid_probs) * x),                          
                     np.sum((y - sigmoid_probs) * 1)]])    
```

```python
def newtons_method(x, y):                                                             
    """
    """

    # Initialize Cost & parameters                                                                   
    theta_1 = 0.001                                                                     
    theta_2 = -0.4 # The intercept term                                                                 
    delta_l = np.Infinity                                                                
    l = Cost(x, y, theta_1, theta_2)                                                                 
    # Convergence Conditions                                                        
    δ = .0000000001                                                                 
    max_iterations = 15                                                            
    i = 0                                                                           
    while abs(delta_l) > δ and i < max_iterations:                                       
        i += 1                                                                      
        g = gradient_nt(x, y, theta_1, theta_2)                                                      
        hess = hessian(x, y, theta_1, theta_2)                                                 
        H_inv = np.linalg.inv(hess)                                                 
        # @ is syntactic sugar for np.dot(H_inv, g.T)¹
        delta = H_inv @ g.T                                                             
        delta_theta_1 = delta[0][0]                                                              
        delta_theta_2 = delta[1][0]  
        print(theta_1,theta_2,l,g)
                                                                                    
        # Perform our update step                                                    
        theta_1 += delta_theta_1                                                                 
        theta_2 += delta_theta_2                                                                 
                                                                                    
        # Update the log-likelihood at each iteration                                     
        l_new = Cost(x, y, theta_1, theta_2)                                                      
        delta_l = l - l_new                                                           
        l = l_new                                                                
    return np.array([theta_1, theta_2])      
```
