## Data Cleaning

```python
# Most people come from Southampton so try fill the na with this value
full_df.Embarked=full_df.Embarked.fillna('S')
```

```python
# Drop Cabin Column: Too much Nan
full_df.drop(['Cabin'], axis = 1, inplace = True)
```

```python
# Fill the age column with the mean age
full_df.Age=full_df.Age.fillna(full_df.Age.mean())
```

```python
full_df.dtypes
```

```python
full_df.columns
```

```python
# Define target 
target = train.Survived
target.shape
```

```python
full_df = full_df[['Age', 'Embarked', 'Fare', 'Parch', 'Pclass', 'Sex', 'SibSp']]
```

```python
# Define training data set
train_df = full_df[:891]
```

```python
train_df.head(3)
```

```python
train_df['Sex'] = train_df['Sex'].map({'male': 1, 'female': 0})
```

```python
train_df.SibSp.value_counts()
```

```python
train_df['Embarked'] = train_df['Embarked'].map({'C':0, 'Q':1, 'S':2})
```

```python
train_df.head(3)
```

这里将age和fare连续型变量转换成了离散型的变量，后面会研究没有转换的模型

```python
#optinal: Filter Age into five groups
def get_age(age):
    if (age <= 12):
        return 0
    elif (12 < age <= 18):
        return 1
    elif (18 < age <= 30):
        return 2
    elif (30 < age <= 60):
        return 3
    elif (60 < age):
        return 4
```

```python
Age = train_df['Age']
```

```python
train_df['Age']= np.array([get_age(x) for x in Age])
```

optinal: Filter Fare into groups

```python
train_df.Fare.describe()
```

```python
def get_fare(fare):
    if (fare <= 8):
        return 0
    elif (8 < fare <= 14):
        return 1
    elif (14 < fare <= 31):
        return 2
    elif (31 < fare):
        return 3
```

```python
Fare = train_df.Fare
```

```python
train_df['Fare'] = np.array([get_fare(x) for x in Fare])
```

```python
train_df.head(3)
```

```python
x = train_df
y = train["Survived"]
```

## Modeling

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2, random_state = 1)
```

```python
print(x_train.shape, x_test.shape)
```

### Building the Model-3 steps:  Define, fit, predict

```python
clf = DecisionTreeClassifier(random_state = 2018)
```

```python
clf.fit(x_train, y_train)
```

```python
predictions = clf.predict(x_test)
```

### Evaluation
[常用的评判标准](https://blog.csdn.net/LFGxiaogang/article/details/82531548)

```python
# ([TN,FN],[FP,TP])混淆矩阵
confusion_matrix(y_test, predictions)
```

```python
accuracy_score(y_test, predictions)
```

## Parameter Tuning

较没有参数限制的结果有所提高精度

```python
predictions_2 = clf_2.predict(x_test)
```
