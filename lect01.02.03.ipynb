# lect03 Python基本数据类型

另外一种对输出格式的定义——利用%s占位符。这是py2的写法，需要在%后面标注类型，py3直接使用{}占位符加format，更简单快捷。

```python
a = 23
w = '70kg'
h = '184cm'
print("I'm %s years old, %s and %s."%(a,w,h) )
print("I'm {} years old, {} and {}.".format(a,w,h))
```

用字符区别不同的占位符，可以重复出现

```python
print("I'm {fir} years old, {sec} and {fir}.".format(fir=a,sec=w,thi=h))
```

还可以用公式来进行占位,在输出字符串的前面加f来区别。

```python
a=5
b=10
print(f'Five plus ten is {a+b} and not {2*(a+b)}.')
```
